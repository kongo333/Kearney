### [참고] 연속형 변수의 구간화 관련 함수 소개

```python
# 조건 만족 여부에 따라 두 가지 값을 갖는 범주형 변수를 생성합니다.
np.where(price['평당금액'].ge(5000), '5천 이상', '5천 미만')
```

```python
# 조건 만족 여부에 따라 세 가지 이상의 값을 갖는 범주형 변수를 생성합니다.
cond = [(price['평당금액'].ge(10000)),
        (price['평당금액'].lt(10000) & price['평당금액'].ge(5000)),
        (price['평당금액'].lt(5000))]
value = ['1억 이상', '5천 이상', '5천 미만']
np.select(condlist = cond, choicelist = value)
```
